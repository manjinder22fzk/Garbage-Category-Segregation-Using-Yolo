**Restart Training**

If the training process was stopped in between either due to Colab's Timeout  or any other issue , you can continue the training from where it stopped . 
For this , you can use the last score of last training . The last score will be stored in ***runs/train/yolov7/garbage_detection/weights/last.pt*** and use the --resume argument

```python
# Only run this code if the training was halted in last code cell
!python train.py --workers 0 --batch-size 4 --device 0 --data data/garbage_detection.yaml --img 640 640 --cfg cfg/training/yolov7_Garbage_Detection.yaml --weights runs/train/yolov7/garbage_detection/weights/last.pt --name yolov7_garbage_detection --hyp data/hyp.scratch.custom.yaml --epochs 300 --resume
```

```python
# Function to Show Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def showImage(path):
  img = mpimg.imread(path)
  plt.figure(figsize=(20,20))
  plt.axis("off")
  plt.imshow(img)
  plt.show()
```
